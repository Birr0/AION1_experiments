```python
import os 
from pathlib import Path 

from dotenv import load_dotenv 
import torch

import matplotlib.pyplot as plt 
import scienceplots
import pandas as pd 
import seaborn as sns 

plt.style.use('science')
```

```python
root = Path(os.getenv("DATA_ROOT"))
folder_name = Path(os.getenv("DATA_FOLDER"))

fp = root / folder_name / "kl_divs_2.0.pt"
```

```python
data = torch.load(fp)

labels = ['True Z', 'Photometry', 'Imaging', 'Spectroscopy', 'Image + Photometry', 'Image + Spectrum', 'Image + Spectrum + Photometry', 'Image + Spectrum + Photometry + Z']
df = pd.DataFrame(
    data=data,
)
df.columns = labels
```

```python
df
```

```python
plt.figure(
    figsize=(10,6),
    dpi=300
)
ax = sns.kdeplot(
    df,
    lw=1.5
)

plt.ylabel(r"$\rho$", fontsize=25)
plt.xlabel(
    r"${D_{\text{KL}}}$",
    fontsize=20
)
plt.xticks(
    fontsize=12
)
plt.xlim(-0.5)
plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
```

```python
plt.figure(
    figsize=(10,6),
    dpi=300
)
ax = sns.histplot(
    df,
    lw=1.5
)

plt.ylabel(r"$\rho$", fontsize=25)
plt.xlabel(
    r"${D_{\text{KL}}}$",
    fontsize=20
)
plt.xticks(
    fontsize=12
)
plt.xlim(-0.5)
plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
```

```python
# plot the delta distributions between each modaility
df["Image + Photometry"] - df["Photometry"]
df["Image + Photometry"] - df["Imaging"]
```
